```python
import pandas as pd
```

```python
dataset = pd.read_csv("Data.csv", parse_dates = ["Date"])
```

```python
dataset.columns = ["Date", "Start", "Visitor", "VPTS", "Home","HPTS", "BS", "Attend"]
```

```python
dataset["HomeWin"] = dataset["VPTS"] < dataset["HPTS"]
dataset["HomeLastWin"] = False
dataset["VisitorLastWin"] = False
```

```python
dataset.loc[::1]
```

```python
y_true = dataset["HomeWin"].values
```

```python
from collections import defaultdict
```

```python
won_last = defaultdict(int)
```

```python
for index, row in dataset.iterrows():
    home_team = row["Home"]
    visitor_team = row["Visitor"]
    row["HomeLastWin"] = won_last[home_team]
    row["VisitorLastWin"] = won_last[visitor_team]
    dataset.iloc[index] = row
    won_last[home_team] = row["HomeWin"]
    won_last[visitor_team] = not row["HomeWin"]
```

```python
dataset.iloc[:40]
```

```python
dataset.columns
```

```python
dataset[:20]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(random_state = 14)
```

```python
X_previouswins = dataset[["HomeLastWin","VisitorLastWin"]].values
```

```python
X_previouswins
```

```python
from sklearn.cross_validation import cross_val_score
scores = cross_val_score(clf, X_previouswins, y_true, scoring = 'accuracy')
```

```python
y_true
```

```python
import numpy as np
print("Accuracy:{0:1f}%".format(np.mean(scores)*100))
```

```python
standing = pd.read_csv("Standings.csv",skiprows = [0])
standing
```

```python
dataset
```

```python
dataset["HomeTeamRanksHigher"] = 0
for index, row in dataset.iterrows():
    home_team = row["Home"]
    visitor_team = row["Visitor"]
    home_rank = standing["Rk"][standing["Team"] ==home_team].values[0]
    visitor_rank = standing[standing["Team"] == visitor_team]["Rk"].values[0]
    row["HomeTeamRanksHigher"] = int(home_rank > visitor_rank)
    dataset.iloc[index] = row
```

```python
dataset
```

```python
X_homehigher = dataset[["HomeLastWin","VisitorLastWin","HomeTeamRanksHigher"]].values
clf = DecisionTreeClassifier(random_state = 14)
scores = cross_val_score(clf,X_homehigher,y_true,scoring = 'accuracy')
```

```python
print("Accuracy:{0:.1f}%".format(np.mean(scores)*100))
```

```python
last_match_winner = defaultdict(int)
dataset["HomeTeamWonLast"] = 0
for index, row in dataset.iterrows():
    home_team = row["Home"]
    visitor_team = row["Visitor"]
    teams = tuple(sorted([home_team,visitor_team]))
    row["HomeTeamWonLast"] = 1 if last_match_winner[teams] == row["Home"] else 0
    dataset.ix[index] = row
    winner = row["Home"] if row["HomeWin"] else row["Visitor"]
    last_match_winner[teams] = winner
```

```python
dataset
```

```python
X_lastwinner = dataset[["HomeTeamRanksHigher", "HomeTeamWonLast"]].values
clf = DecisionTreeClassifier(random_state = 14)
scores = cross_val_score(clf, X_lastwinner, y_true, scoring = 'accuracy')
```

```python
print("Accuracy:{0:.1f}%".format(np.mean(scores)*100))
```
